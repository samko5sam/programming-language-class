```python
sheetData = gsheets.get_worksheet(0)
data = sheetData.get_all_values()  # 取得所有儲存格資料，回傳為 2D 陣列
df_money = pd.DataFrame(data[1:], columns=data[0])
df_money
```

```python
from IPython.display import Markdown
mdTable = df_money.to_markdown()
prompt = f"這是ETF股票0050的除權息資料，請對資料進行摘要，再根據資料給股票新手一些建議：\n{mdTable}"
response = model.generate_content(prompt)
display(Markdown(response.text))
```

```python
%%capture
from bs4 import BeautifulSoup
from markdown import markdown
import re

def markdown_to_text(markdown_string):
    """ Converts a markdown string to plaintext, preserving list markers """

    # Convert markdown to HTML
    html = markdown(markdown_string)

    # Remove code blocks
    html = re.sub(r'<pre>.*?</pre>', ' ', html, flags=re.DOTALL)
    html = re.sub(r'<code>.*?</code>', ' ', html, flags=re.DOTALL)

    # Parse HTML with BeautifulSoup
    soup = BeautifulSoup(html, "html.parser")

    # Preserve list markers
    text_lines = []
    for element in soup.find_all(["p", "li"]):
        if element.name == "li":
            # Check if it's inside an ordered (`ol`) or unordered (`ul`) list
            parent = element.find_parent(["ul", "ol"])
            if parent and parent.name == "ul":
                text_lines.append(f"    * {element.get_text(strip=True)}")
            elif parent and parent.name == "ol":
                # Get the index of the list item
                index = parent.find_all("li").index(element) + 1
                text_lines.append(f"    {index}. {element.get_text(strip=True)}")
        else:
            text_lines.append(element.get_text(strip=True))

    return "\n".join(text_lines)

sheetOut = gsheets.get_worksheet(1)
plain_text = markdown_to_text(response.text)
sheetOut.update_acell('A1', plain_text)
```
